```python
#from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report,accuracy_score 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
import statistics
```

```python
df=pd.read_csv('MSFT_groupby_all.csv')
df = df.fillna('N') #populating each player's last round empty cells with letter N
```

```python
statistics.stdev(df["open"])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

rf = RandomForestRegressor()

df_target = df["open"]

df_features = df
df_features = df_features.drop(["open"],axis=1)
df_features = df_features.drop(["date"],axis=1)



n_train = 1
n_records = len(df)
total_count=0
totacc=0
ytlog=[]
yplog=[]
pred=[df_target[0]]
for i in range(n_train, n_records):
    X_train, X_test = df_features[0:i],df_features[i:i+1]
    y_train, y_test = df_target[0:i], df_target[i:i+1]
    rf.fit(X_train.values, y_train)
    y_pred = rf.predict(X_test.values)
    pred.append(y_pred)

print(mean_squared_error(df["open"], pred)**(1/2))

'''plt.figure()
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.plot(df["date"], df["open"],pred)
plt.show()'''

fig, ax = plt.subplots(1,1)
#plt.rcParams["figure.figsize"] = (6,1)
plt.title("Microsoft")
#plt.xlabel("Date")
plt.ylabel("Stock Price")
ax.plot(df["date"],df["open"],pred)
ax.xaxis.set_major_locator(ticker.MultipleLocator(50))
plt.show()
```
